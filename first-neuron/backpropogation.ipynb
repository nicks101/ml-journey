```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 3*x**2 - 4*x + 5
```

```python
f(3.0)
```

```python
xs = np.arange(-6, 6, 0.25)
ys = f(xs)
```

```python
plt.plot(xs, ys)
```

```python
h = 0.001
x = 0.7

# derivative/slope
slope = (f(x + h) - f(x)) / h
print('At', x, 'y=', f(x))
print('At', x+h, 'y=', f(x+h))

#slope ((y - y0) / (x - x0) )
# tangent = y values of slope
tangent = f(x) + slope*(xs - x) 

plt.plot(xs, ys, label='f(x)')
plt.plot(xs, tangent, label='slope')

# By changing the x at a particular point, how much is function changing and is the function increasing or decreasing
```

```python
# To keep track of more complex equation and derivatives

class Value:

    def __init__(self, data, _children=()):
        self.data = data
        self._prev = set(_children)

    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        r = Value(self.data + other.data, (self, other))
        return r

    def __mul__(self, other):
        r = Value(self.data * other.data, (self, other))
        return r

a = Value(2.0)
b = Value(-3.0)
c = Value(10.0)
e = a * b
d = e + c
```
